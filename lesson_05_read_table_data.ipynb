```python
# DataFrame를 만들때 컬럼명과 인텍스명을 지정
data = [
    [60, 65, 66],
    [80, 85, 88],
    [100, 100, 100]
]

col = ["국어", "수학", "영어"]
idx = ["가영", "나미", "다윤"]
df = pd.DataFrame(data, columns=col, index=idx)
df
```

```python
data = {
    "국어": [60, 80, 100],
    "수학": [65, 85, 100],
    "영어": [66, 88, 100]
}
idx = ["가영", "나미", "다윤"]
df= pd.DataFrame(data, index=idx)
df
```
